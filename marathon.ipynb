season relationship with average speed
spring 3-5
summer 6-8
fall 9-11
winter 12-2

```python
df3['race_month'] = df3['race_day'].str.split('.').str.get(1).astype(int)
```

```python
df3
```

```python
df3['race_season'] = df3['race_month'].apply(lambda x: 'Winter' if x > 11 else 'Fall' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'Winter')
```

This lambda function essentially maps each month to its corresponding season based on the conditions specified. It's a concise way to categorize data into different groups based on certain rules or conditions

,

```python
df3.query('race_season == "Summer"')
```

```python
df3.groupby('race_season')['athelete_avg_speed'].agg(['mean','count']).sort_values('mean',ascending = False)
```
